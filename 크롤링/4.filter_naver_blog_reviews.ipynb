```python
def contains_any_keyword(text, keyword_list):
    return any(keyword in text for keyword in keyword_list)
```

```python
# ì˜ˆì‹œ í‚¤ì›Œë“œ (ì›í•˜ëŠ” ëŒ€ë¡œ ìˆ˜ì • ê°€ëŠ¥)
# irrelevant_keywords = [
#     "ë§›ì§‘", "ì¹´í˜", "ë ˆìŠ¤í† ë‘", "ì‹ë‹¹", "ì»¤í”¼", "ë””ì €íŠ¸",
#     "ì„ëŒ€", "ë¶€ë™ì‚°", "ì‚¬ë¬´ì‹¤", "ì›”ì„¸", "ìƒê°€",
#     "ì†ëˆˆì¹", "ì™ì‹±", "í”¼ë¶€", "ì˜ë£Œ", "í´ë¦¬ë‹‰",
#     "ë§¤ë¬¼", "ìˆ ì§‘", "ë·°í‹°", "ë„¤ì¼", "ë³‘ì›", "í”¼ë¶€ê³¼",
#     "í—¬ìŠ¤ì¥", "í—¬ìŠ¤", "í—¬ìŠ¤í´ëŸ½", "ìŠ¤íŒŒ", "ë§ˆì‚¬ì§€",
#     "ë¶„ì–‘", "ì•„íŒŒíŠ¸", "ì˜¤í”¼ìŠ¤í…”", "ë¶„ì–‘ê¶Œ",
#     "ì˜·", "ê°€ë°©", "ê´‘ê³ ", "í˜‘ì°¬", "ì´ë²¤íŠ¸", "ì²´í—˜ë‹¨", "ë°°ì†¡", "ì œí’ˆ", 
#     "ë¦¬ë·°ì–´", "ì‡¼í•‘ëª°", "êµ¬ë§¤", "ë¸Œëœë“œ", "íŒ¨ì…˜", "í™”ì¥í’ˆ", 
#     "ë·°í‹°", "ìŠ¤í‚¨ì¼€ì–´", "ë§ˆìŠ¤í¬íŒ©", "íƒë°°", "ì œí’ˆí›„ê¸°", "í• ì¸"
# ]
strict_irrelevant_keywords = [
    "ì„ëŒ€", "ë¶€ë™ì‚°", "ì‚¬ë¬´ì‹¤", "ì›”ì„¸", "ìƒê°€",
    "ì™ì‹±", "ì˜ë£Œ", "í´ë¦¬ë‹‰",
    "ë¶„ì–‘", "ì•„íŒŒíŠ¸", "ì˜¤í”¼ìŠ¤í…”", "ë¶„ì–‘ê¶Œ",
    "íƒë°°", "ì‡¼í•‘ëª°", "ë¸Œëœë“œ", "ë¦¬ë·°ì–´", "ì²´í—˜ë‹¨"
]
```

```python
# ê´€ë ¨ ì—†ëŠ” í–‰ ì œê±°
filtered_df = df[~df["content"].apply(lambda x: contains_any_keyword(str(x), strict_irrelevant_keywords))]

# ê²°ê³¼ í™•ì¸
print(f"í•„í„°ë§ ì „: {len(df)}ê°œ â†’ í•„í„°ë§ í›„: {len(filtered_df)}ê°œ")
```
