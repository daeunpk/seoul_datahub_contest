```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# 1. 데이터 불러오기
file_path = "../../data/MED-MEDINST/MED-MEDINST.csv"
df = pd.read_csv(file_path)

# 2. '소계' 제거
df = df[df["자치구"] != "소계"].reset_index(drop=True)

# 3. 숫자형 변환 (에러시 NaN 처리)
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# 4. 총 병원수, 총 병상수 계산
df["hospital_count"] = df["병원수 소계"]
df["bed_count"] = df["병상수 소계"]

# 5. 점수화 (0~100)
scaler = MinMaxScaler()
df[["hospital_count_score", "bed_count_score"]] = scaler.fit_transform(
    df[["hospital_count", "bed_count"]]
) * 100

# 6. 평균 점수
df["medinst_mean_score"] = df[["hospital_count_score", "bed_count_score"]].mean(axis=1)

# 7. 필요한 컬럼만 정리
result = df[["자치구", "hospital_count_score", "bed_count_score", "medinst_mean_score"]].rename(
    columns={"자치구": "gu"}
)

# 8. 저장
result.to_csv("../../result/MED/MED-MEDINST_result.csv", index=False, encoding="utf-8-sig")


result.head()
```
